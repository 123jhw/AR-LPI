```python
from scipy.fft import dct
import numpy as np
import RNA
```

```python

def stringsToVector(texts):
    vectors = []
    for text in texts:
        numberArray = []
        for i in text:
            numberArray.append(int(i))
        vectors.append(np.array(numberArray))
    return vectors
```

```python

def vectorFourier(vectors):
    result = vectors[0]
    for i in range(1, len(vectors)):
        result += vectors[i]
    return dct(result, type=4, norm='ortho')
```

```python
RNA.cvar.uniq_ML = 1
RNA_str = []
print("序列超长的RNA有: ")
for i in range(0,len(data_seq1807_XR)):
    sequence = data_seq1807_XR_new.iloc[i,1]
    print("proccessing #" + str(i) + " RNA:" + sequence)

    structure = RNA.fold(sequence)

    
    solution = structure[0]
    
    texts=solution.replace("(", "1").replace(")", "1").replace(".", "0")

    result = vectorFourier(stringsToVector([texts]))[:10]
    RNA_str.append(result)
    print("#" + str(i) + " RNA fourier transform's result is " + str(result))
    print("\n")
```
